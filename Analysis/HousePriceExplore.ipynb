Exploring the House Price Data
==============================

Let's take a look at the data:

```python
fig, ax = plt.subplots()
hp["Price"].hist(ax=ax, bins=100, bottom=1)
ax.set_yscale('log')
plt.show(fig)
```

```python
fig, ax = plt.subplots()
maxP = hp["Price"].max()
minP = hp["Price"].min()
hp["Price"].hist(ax=ax, bins=10**np.linspace(np.log10(minP), np.log10(maxP),100), bottom=1)
ax.set_yscale('log')
ax.set_xscale('log')
plt.show(fig)
```

That's some tasty distribution. New question: how do the house types e have access to affect the price?

```python
hp.boxplot(column="Price", by ="Property_Type", sym="");
hp.boxplot(column="Price", by ="Freeorlease", sym="");
hp.boxplot(column="Price", by ="Newbuild", sym="");
```
