```python
fig, axs = plt.subplots(4)
maxP = hp["Price"].max()
minP = hp["Price"].min()
Pbins = 10**np.linspace(np.log10(minP), np.log10(maxP),100)
propTypes = np.unique(hp["Property_Type"].values)
for n,ax in enumerate(axs):
    ax.hist(hp[hp["Property_Type"] == propTypes[n]]["Price"].values, bottom = 10**(-11), bins=Pbins, normed=True)
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.set_title(propTypes[n])
fig.subplots_adjust(hspace=0.5)
fig.set_size_inches(5,10)
plt.show(fig)
```

These graphs are all plotted in log-log space, which tells us that that linear slop present in all of the graphs tells us that the tails of the distribution follow a power law:

\begin{equation}
p(y) \propto x^{a}
\end{equation}

Exploration part 2: As a function of distance
=============================================

What about houseprice as a function of position relative to the center of london?

```python
hpg = hp.groupby("Lsoa11").agg({"Price" : np.median, 
                                "Newbuild" : (lambda x : (x == "Y").sum())})
```

```python
pc.rename(columns = {"ZONEID":"Lsoa11"}, inplace=True)
Zonelocation = pc[ ["Lsoa11","Latitude", "Longitude"] ].groupby("Lsoa11").mean()
```

```python
test = Zonelocation.join(hpg)
```

```python
#Find the center of london
#Turns out to be somewhere on the strand, close to temple station
cx, cy = test["Longitude"].median(), test["Latitude"].median()
```

```python
def longLatDist(p1, p2): 
    '''Takes two pairs of (long,lat) and returns distance between them (using small angle approx)'''
    R = 6371 #Earth Radius in kilometers
    dx = R*np.cos(np.radians(p1[0]))*np.radians(p1[0] - p2[0])
    dy = R*np.radians(p1[1] - p2[1])
    return np.sqrt( dx**2 + dy**2 )

def getDist(r):
    return longLatDist( (r["Longitude"], r["Latitude"]), (cx, cy) )
```

```python
test["dist"] = test.apply(getDist, axis=1)
test["logPrice"] = test["Price"].apply(np.log10)
```

```python
test["dist"].hist(bins=100);
```

Not what you'd expect for a uniformly distributed population on a circle, but that's not what we're dealing with. London's population density drop as you go out from the center, which is reflected in LSOA size.

```python
test.plot(y = "Price", x = "dist", style="o");
```

```python
test.plot(y = "logPrice", x = "dist", style="o");
```

There is an effect, but the whole thing seems to be skewed by a few really expensive areas.
