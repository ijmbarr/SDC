```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import json
#%matplotlib inline
```

```python
mapDirectory = "../Data/MapData/"
hpDirectory = "../Data/HousePriceData/"
pcDirectory = "../Data/PostCodeData/"
```

```python
# Load Files
hp = pd.read_csv(hpDirectory + "HousePriceReduced.csv")
del hp["Unnamed: 0"]

pc = pd.read_csv(pcDirectory + "postCodes.csv")

f = open(mapDirectory + "LSOA.json","r")
lsoaMap = json.load(f)
f.close()

f = open(mapDirectory + "borough.json","r")
boroughMap = json.load(f)
f.close()
```

```python
#Functions for plotting geojson features
from matplotlib.patches import PathPatch
from matplotlib.path import Path

#codes for creating the polygons
pathCodes = [Path.MOVETO, Path.LINETO, Path.CLOSEPOLY]

def toCart(coords): 
    #Converts (long,lat) lists to cartesian co-ordinates, assuming small angle approx
    return map( lambda x : [np.cos(np.radians(x[1]))* x[0], x[1]], coords )

def drawFeature(feature, color = "none", lineW=0.5):
    path = []
    codes = []
    
    if feature["geometry"]["type"] == "Polygon":
        path, codes = drawPoly(feature["geometry"]["coordinates"])
        
    elif feature["geometry"]["type"] == "MultiPolygon":
        for f in feature["geometry"]["coordinates"]:
            temp = drawPoly(f)
            path += temp[0]
            codes += temp[1]        
    else:
        path = None
        
    return PathPatch(Path(path, codes), lw=lineW, fc = color)

    
def drawPoly(coords):
    path = []
    codes = []
    for f in coords:
        path += toCart(f)
        codes += [pathCodes[0]] + [pathCodes[1]]*(len(f) - 2) + [pathCodes[2]]
    
    return (path,codes)
```

```python
#Plot London's boroughs
border = 0.3
cx, cy =  -0.05, 51.5

fig, ax = plt.subplots()

for x in boroughMap["features"]:
    patch = drawFeature(x)
    ax.add_patch(patch)
    
ax.set_xlim(cx-border, cx+border)
ax.set_ylim(cy-border, cy+border)

plt.show()
```

```python
def addPoints(ax, data):
    col = np.log10(data["Price"].values)
    x = np.asarray(toCart(data[ ["Longitude","Latitude"] ].values))
    sc = ax.scatter(x[:,0], x[:,1], c=col, marker=".", cmap = plt.cm.hsv, lw=0, vmin=4.0,vmax=7.2) 
    plt.colorbar(sc)
```

```python
#Visulise all the sales over 4 years by month
border = 0.3
cx, cy =  -0.05, 51.5

fig, ax = plt.subplots()

plt.ion()

months = np.unique(hp["Month"].values)

for m in months[:]:
    plt.clf()
    ax = fig.add_subplot(111)
    
    for f in boroughMap["features"]:
        patch = drawFeature(f)
        ax.add_patch(patch)
        
    ax.set_xlim(cx-border, cx+border)
    ax.set_ylim(cy-border, cy+border)
    
    ax.set_title("Month: {}".format(m))
    addPoints(ax, hp[hp["Month"] == m])
    
    plt.draw()

plt.ioff()
plt.show()
```

We can now visulise the same thing, but averaging house price over each given LSOA:

```python
#Visulise median price by LSAO
import matplotlib

border = 0.3
cx, cy =  -0.05, 51.5

fig, ax = plt.subplots()

converter = hp.groupby("Lsoa11")["Price"].median()

#Make a color map:
minP = np.log10(converter.min())
maxP = np.log10(converter.max())
cNorm = matplotlib.colors.Normalize(vmin = minP, vmax = maxP)
cm = matplotlib.cm.ScalarMappable(norm=cNorm, cmap = plt.cm.spectral)
cm._A =[]

#Add each LSOA to the plot
for f in lsoaMap["features"][:]:
    lsoa = f["properties"]["LSOA11CD"]
    data = converter.loc[lsoa]
    col = cm.to_rgba(np.log10(data))
    patch = drawFeature(f, col, lineW=0)
    ax.add_patch(patch)
    
ax.set_xlim(cx-border, cx+border)
ax.set_ylim(cy-border, cy+border)

plt.colorbar(cm)

plt.show()
```

```python
#Visulise something by LSAO
import matplotlib

border = 0.3
cx, cy =  -0.05, 51.5

fig, ax = plt.subplots()

def f(a): return (a["Newbuild"] == "Y").sum()
converter = hp.groupby("Lsoa11").apply(f)

#Make a color map:
minP = converter.min()
maxP = converter.max()
cNorm = matplotlib.colors.Normalize(vmin = minP, vmax = maxP)
cm = matplotlib.cm.ScalarMappable(norm=cNorm, cmap = plt.cm.spectral)
cm._A =[]

for f in lsoaMap["features"][:]:
    lsoa = f["properties"]["LSOA11CD"]
    data = converter.loc[lsoa]
    col = cm.to_rgba(data)
    patch = drawFeature(f, col, lineW=0.1)
    ax.add_patch(patch)
    
ax.set_xlim(cx-border, cx+border)
ax.set_ylim(cy-border, cy+border)

plt.colorbar(cm)

plt.show()
```
